```python
import qiskit
from qiskit import QuantumCircuit, QuantumRegister
from random import seed
from random import random
import numpy as np
```

```python
num_qubit = 5
depth = 3
register = QuantumRegister(num_qubit)
circuit = QuantumCircuit(register)
params = []
for i in range(depth):
    for j in range(num_qubit):
        rx1 = random()
        rz1 = random()
        rx2 = random()
        params.append(rx1)
        params.append(rz1)
        params.append(rx2)
        circuit.rx(rx1, register[j])
        circuit.rz(rz1, register[j])
        circuit.rx(rx2, register[j])
        cx_counter = 1
    while cx_counter <= len(register) - 1:
        circuit.cx(register[cx_counter - 1], register[cx_counter])
        cx_counter += 1
    circuit.barrier(register)
params = np.array(params)
circuit.draw(scale=0.05)
```
