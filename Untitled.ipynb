```python
import qiskit
from qiskit import QuantumCircuit, QuantumRegister
from random import seed
from random import random
```

```python
num_qubit = 10
depth = 9
register = QuantumRegister(num_qubit)
circuit = QuantumCircuit(register)
for i in range(num_qubit):
    circuit.rx(random(), register)
    circuit.rz(random(), register)
    circuit.rx(random(), register)
    cx_counter = 1
    while cx_counter <= len(register) - 1:
        circuit.cx(register[cx_counter - 1], register[cx_counter])
        cx_counter += 1
print(circuit)
```
